## 3. Megapode

### Megapodes or Brush Turkeys lay their eggs and then cover them with a giant pile of dirt and brush. Rather than incubate in the classic style, they have some kind of sixth sense that allows them to add and remove organic matter from the pile to regulate the temperature of the eggs. They do this faithfully during the entire incubation phase until hearing the peeps of the about-to-hatch turkey poults. Then they literally just walk away. The poults hatch from the eggs, dig themselves out, and they're on their own. SO PUNK.

## 1. Lammergeier

### Lammergeiers or Bearded Vultures eat bones. That's it, just bones. Not the flesh off the bones. Bones. They drop e.g. goat femurs from high heights in the Ethiopian mountains, shattering them on the rocks below, then picking up (still large) bone bits and swallowing them whole. They have evolved special digestive juices so acidic they dissolve the bone quickly, making for a nutritious meal of BONES. Also they paint themselves with red mud to make themselves look more badass. As if this were necessary.
